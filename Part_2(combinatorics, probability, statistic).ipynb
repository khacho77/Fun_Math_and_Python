ხშირად საჭიროა სასრულ ელემენტებისგან სხვადასხვა კომბინაციის შედგენა და რაიმე წესით შედგენილი ყველა შესაძლო კომბინაციის რაოდენობის გამოანგარიშება.

###### გადანაცვლება. 
ერთიდაიგივე სასრული სიმრავლის ელემენტები შეიძლება დავალაგოთ რიგით იმის მიხედვით, თუ სიმრავლის რომელ ელემენტს ავირჩევთ პირველ ელემენტად, რომელს-მეორედ და ა.შ. დალაგებული სიმრავლის აღსანიშნავად მის ელემენტებს მოვათავსებთ მრგვალ ფრჩხილებში მოცემული რიგი მიხედვით.

სასრულ სიმრავლეში დადგენილ რიგს მისი ელემენტების გადანაცვლება ეწოდება.

n ელემენტიანი სიმრავლის ყველა შესაძლო გადანაცვლებათა რიცხვი

ცხადია იგი დამოკიდებულია მხოლოდ სიმრავლის ელემენტთა რაოდენობაზე.

n ელემენტიანი სიმრავლის გადანაცვლებათა რიცხვის გამოსათვლელი ფორმულა შეიძლენა შემდეგნაირად ჩაიწეროს:

```python
import math

n = 5

math.factorial(n)
```

###### წყობა.  
n ელემენტიანი სიმარავლის ნებისმიერ m ელემენტიან დალაგებულ ქვესიმრავლეს (m ≤ n) ეწოდება წყობა n ელემენტისგან m-ად.

n ელემენატიანი სიმარავლის ყველა შესაძლო m ელემენტიან წყობათა რიცხვი სიმბოლოთი აღინიშნება.


მოცემული n ელემენტისგან მ ელემენატიანი წყობის მისაღები ფორმულაა:

```python
import math

n = 4
m = 2

math.factorial(n) / math.factorial(n - m)
```

###### ჯუფთება.
n ელემენტიანი სიმრავლის ნებისმიერ m ელემენტიან ქვესიმრავლეს (m ≤ n) ეწოდება ჯუფთება n ელემენტიდან
m - ად.
ცხადია, რომ მოცემული სიმრავლის ორი m  ელემენტიანი ჯუფთება სხვადასხვაა მაშინ და მხოლოდ მაშინ, როდესაც
ისინი განსხვავდებიან ერთი ელემენტით მაინც, ხოლო m ელემენტიანი წყობები რომ განსხვავდებოდნენ, საკმარისია ისინი განსახვავდებოდნენ დალაგებით.

n ელემენტიანი სიმრავლის ყველა შესაძლო m ელემენტიან ჯუფთებათა რიცხვის სიმბოლოთი აღინიშნება.



მოცემული n ელემენტისგან შედგენილი ყველა შესაძლო m ელემენტიანი წყობის მისაღები ფორმულაა:

```python
from math import factorial

def choose(n, c):
    "n ელემენტის სიიდან m ელემენტის არჩევის გზების რაოდენობა."
    return factorial(n) // (factorial(n - c) * factorial(c))

choose(4, 2)
```

Isaac Newton; დ. 25 დეკემბერი, 1642 (4 იანვარი, 1643), სულსთორპი, გრემთემის მახლობლად — გ. 20 (31) მარტი, 1727, კენსინგტონი) — ინგლისელი ფიზიკოსი, მათემატიკოსი, ასტრონომი, ალქიმიკოსი და ფილოსოფოსი; კავშირშია მეცნიერულ რევოლუციასა და ჰელიოცენტრიზმის განვითარებასთან. ერთ-ერთი უმნიშვნელოვანესი, გავლენიანი მეცნიერი ისტორიაში.

###### პასკალის სამკუთხედი

1653 წელს ფრანგმა მათემატიკოსმა და ფილოსოფოსმა ბლეზ პასკალმა მონეტის აგდების შედეგების შესაბამისი რიცხვების სამკუთხა კონფიგურაცია აღწერა. შევნიშნავთ, რომ ეს იგივეა, რაც ცალი ობიექტიდან ცალი ობიექტის შესაძლო ამორჩევათა რაოდენობა. მიუხედავად იმისა, რომ პასკალს ამ აღმოჩენაზე პრეტენზია არ გამოუთქვამს, მისი სახელი განუყრელად დაუკავშირდა ამ ფაქტს და რიცხვების ამ სამკუთხა კონფიგურაციას პასკალის სამკუთხედი ეწოდა. სინამდვილეში ეს სამკუთხედი რამდენიმე საუკუნით ადრე აღიწერა. პირველად იგი იხსენიება ინდოელი მათემატიკოსის პინგალას წიგნში, რომელიც სანსკრიტულ პოეზიას ეძღვნება და შეიქმნა ჩვენს წელთაღრიცხვამდე დაახლოებით 450 წელს. ასევე ცნობილია, რომ ამ სამკუთხედის შესახებ იცოდნენ ძველმა ჩინელმა მათემატიკოსებმა. მოგვიანებით სპარსმა მათემატიკოსმა ყარაჯიმ და პოეტმა- ასტრონომმა ომარ ხაიამმა იგი მოიხსენიეს როგორც „ხაიამის სამკუთხედი“. ცნობილი იყო მასთან დაკავშირებული რამდენიმე თეორემა, მათ შორის - ე.წ. ბინომური თეორემა. იტალიაში ამ სამკუთხედს -ს (ტარტალიას სამკუთხედს) უწოდებდნენ იტალიელი მათემატიკოსის ნიკოლო ფონტანა ტარტალიას პატივსაცემად.

პასკალის სამკუთხედის თვისებაა რომ მის შიგნით ნებისმიერი რიცხვი უდრის მის თავზე მდგომი ორი რიცხვის ჯამს და შეესაბამება ტოლობას:



რიცხვებს ეწოდება ბინომური კოეფიციენტები, ხოლო ბინომური განაწილება ეწოდება ალბათობებს:



ბლეზ პასკალი (ფრანგ. Blaise Pascal; დ. 19 ივნისი, 1623, კლერმონ-ფერანი — გ. 19 აგვისტო, 1662, პარიზი) — ფრანგი მათემატიკოსი, ფიზიკოსი, ლიტერატორი, ფილოსოფოსი, მორალისტი, კიბერნეტიკოსი, ალბათობის თეორიის ერთ-ერთი ფუძემდებელი, ჰიდროსტატიკის, როგორც მეცნიერების ერთ-ერთი დამაარსებელი.


```python
# Print Pascal's Triangle in Python
from math import factorial

# input n
n = 6
for i in range(n+1):
    for j in range(n-i+1):
         print(end=" ")
 
    for j in range(i+1):
         print(factorial(i)//(factorial(j) * factorial(i-j)), end=" ")
    print()
```

## ალბათობის თეორიის ელემენტები.

1814 წელს პიერ-სიმონ ლაპლასმა დაწერა:

ალბათობა არის უბრალოდ წილადი, რომლის მრიცხველი არის ხელსაყრელი შემთხვევების რაოდენობა და რომლის მნიშვნელი არის ყველა შესაძლო შემთხვევის რიცხვი, როცა არაფერი გვაიძულებს ვივარაუდოთ, რომ ამ შემთხვევიდან რომელიმე სხვაზე მეტად უნდა მოხდეს.

პიერ სიმონ ლაპლასი (დ. 23 მარტი, 1749, ბომონ-ან-ოჟი, ნორმანდია ― გ. 5 მარტი, 1827, პარიზი) — ფრანგი ასტრონომი, მათემატიკოსი და ფიზიკოსი, პარიზის მეცნიერებათა აკადემიის წევრი (1785, ადიუნქტი 1773–დან), საფრანგეთის მეცნიერებათა აკადემიის წევრი (1816).





ალბათობის თეორია არის მეცნიერება, რომელიც შეისწავლის მასობრივ შემთხვევით მოვლენათა რაოდენობრივი ხასიათის კანონზომიერებებს.

ალბათობის თეორიის პირველად ცნებას ხდომილება წარმოადგენს.

მაგალითად, მონეტის აგდებისას შესაძლო შედეგი ორია: გერბი და საფასური. საფასური შეიძლება მოვიდეს ან შეიძლება არც მოვიდეს. ასეთ შემთხვევაში ამბობენ საფასურის მოსვლა შემთხვევითი ხდომილებაა. შემთხვევითი ხდომილებაა მაგალითად: კამათელზე კენტი რიცხვის მოსვლა. ამრიგად ხდომილება შემთხვევითია თუ მას ერთსა და იმავე პირობებში შეიძლება ჰქონდეს ან არ ჰქონდეს ადგილი.

ზოგი ხდომილება ისეთია, რომ მას აუცილებლად ექნება ადგილი, თუ ცდას ჩავატარებთ , ასეთ ხდომილებას აუცილებელ ხდომილებას უწოდებენ.

ხდომილებას, რომელსაც არ შეიძლება ჰქონდეს ადგილი რაიმე ცდის ჩასატარების დროს, შეუძლებელი ხდომილება ეწოდება.

ერთი კამათლის გაგორებისას წინასწარ შეუძლებელია თქმა, რა რიცხვი მოვა, მაგრამ შეიძლება ვივარაუდოთ, რომ თითოეული რიცხვის მოსვლა თანაბრადაა მოსალოდნელი. ასეთ შესაძლო შედეგებს თანაბრად მოსალოდნელი ეწოდება.

```python
import numpy as np

def kubik(n):
    """
    n: გაგორების რაოდენობა
 
    """
    data = []
    while len(data) < n:
        data.append(np.random.randint(1,7))
    return data

kubik(10)
```

###### ოპერაციები ხდომილებებზე

ოპერაციები ხდომილებებზე ისე განისაზღვრება , როგორც სიმრავლეზე ოპერაციები.

ორი A და B ხდომილების ჯამი(გაერთიანება) ეწოდება ისეთ ხდომილებად, რომელიც ხდება მაშინ და მხოლოდ მაშინ როდესაც ხდება ერთი მაინც A და B ხდომილებიდან. აღინიშნება ასე A+B ან A⋃B.

თუ A და B თავსებადი P(A+B)<P(A)+P(B).
თუ A და B არათავსებადია P(A+B)=P(A)+P(B).

ორი A და B  ხდომილების ნამრავლი(თანაკვეთა) ეწოდება ისეთ ხდომილებას, როიმელიც ხდება მაშინ და მხოლოდ მაშინ, როდესაც ხდება როგორც A, ისე  B  ხდომილება. აღინიშნება ასე AB ან A⋂B.

თუ ადგილი აქვს ტოლობას P(A⋅B) = P(A)⋅P(B), მაშინ A და B ხდომილებებს დამოუკიდებელი ხდომილებები ეწოდება.წინააღმდეგ შემთხვევაში ამ ხდომილებებს დამოკიდებულ ხდომილებები ეწოდება. 

თუ A და B ნებისმიერი ხდომილებებია, მაშინ P(A+B)=P(A)+P(B)−P(A⋅B).

მაგალითად:

ვთქვათ ქსპერიმენტია კამათლის გაგორება A იყოს  ხდომილება "4-ზე მეტი რიცხვის მოსვლა". B იყოს ხდომილება - "3-ზე მეტი და 6-ზე ნაკლები რიცხვის მოსვლა", ამიტომ A + B = {4,5,6}. ცხადია რომ P(A) = 1/3 ; P(B) = 1/3,
ხოლო P(A+B) = 1/2. P(A⋅B) = 1/6 . აღნუშნული მაგალითიდან მოწმდება ტოლობის მართებულობა.


```python
A = {5,6} 
B = {4,5}
print(A | B)   # P(A+B) = 1/2
print(A & B)   # P(A*B) = 1/6
```

```python
# P(A+B)=P(A)+P(B)−P(A⋅B).

import fractions as F

F.Fraction("1/2") == F.Fraction("1/3")+F.Fraction("1/3")-F.Fraction("1/6")
```

ორი A და B  ხდომილების სხვაობა ეწოდება ისეთ ხდომილებას, რომელიც ხდება მაშინ და მხოლოდ მაშინ, როდესაც ხდება A და არ ხდება B. აღინიშნება A − B ან A\B.

საწინააღმდეგო ხდომილობის ალბათობის გამოთვლა: P(A)=1−P(A).
