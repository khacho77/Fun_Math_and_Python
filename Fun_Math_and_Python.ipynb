• რამდენიმე რიცხვის უდიდესი საერთო გამყოფისა და უმცირესი საერთო ჯერადის პოვნა.

ნატურალური რიცხვი — მთელი დადებითი რიცხვი (1, 2, 3, 4, ...) . ნატურალურ რიცხვებს უწოდებენ რიცხვებს, რომლებიც საგნების დასათვლელად გამოიყენება ნატურალურ რიცხვს ეწოდება მარტივი, თუ მას მხოლოდ ორი გამყოფი (თავის თავი და ერთი) აქვს, ხოლო შედგენილი, თუ ორზე მეტი გამყოფი აქვს.

```python
n = 55   # ჩაწერეთ ნატურალური რიცხვი

try:
    if n > 0:
        total = 0
        for i in range(1, n + 1):
            if n % i == 0:
                total += 1
        if n == 1 :
                print("მარტივი")
        elif total == 2:
            print("მარტივი")
        else:
            print("შედგენილი")
    else:
        print("ჩაწერეთ ნატურალური რიცხვი")
except:
    print("ჩაწერეთ ნატურალური რიცხვი")
```

### 3.  მთელი რიცხვები.

• არითმეტიკული მოქმედებები მთელ რიცხვებზე.

დადებითი რიცხვი - 0 ზე მეტია და იწერება „+“ ნიშნით ან ვწერთ რიცხვს ნიშნის გარეშე

უარყოფითი რიცხვი - არის დადებითი რიცხვის მოპირდაპირე და იწერება მინუს „ – “ ნიშნით. უარყოფითი რიცხვი 0-ზე ნაკლებია.

ნატურალურ რიცხვებს მათ მოპირდაპირე რიცხვებს და 0-ს მთელი რიცხვები ეწოდება.

ნული — რიცხვი, რომელთანაც შეკრებისას ნებისმიერი რიცხვი უცვლელი რჩება. ნული აღინიშნება სიმბოლოთი 0. ნებისმიერი რიცხვის ნამრავლი ნულზე უდრის ნულს. a * 0 = 0. ნულზე გაყოფა მათემატიკაში განსაზღვრული არ არის და ყოველგვარ აზრსაა მოკლებული

გამრავლების ტაბულა

```python
for i in range(17, 19):

    for j in range(17, 19):

        print(f'{i} x {j} = {i * j}')

    print()
```

 ფიბონაჩის რიცხვების თანმიმდევრობის ყოველი რიცხვის ფარდობა მის შემდგომ რიცხვთან დიდი მიახლოებით ტოლია 0.618-ის აი მაგალითად 55/89 დაახლოებით 0.618-ის ტოლია, ხოლო ყოველი რიცხვის ფარდობა მის წინა რიცხვთან დიდი მიახლოებით ტოლია 1.618-ის ანუ 0.618-ის ინვერსიის. რაც უფრო იზრდება რიცხვები მით უფრო ზუსტად უახლოვდება ფარდობა 0.618034…-ს რიცხვს რომელსაც ვუწოდებთ φ-ს (ფი). 
 
 ფიბონაჩის რიცხვებს ხშირად ბუნების ნუმერაციის სისტემასაც უწოდებენ , რადგან სამყაროში თითქმის ყველაფერი მასა და ოქროს კვეთის პროპორციას ემყარება

```python
# პროგრამა ფიბონაჩის მიმდევრობის აჩვენების n - რიცხვამდე
 
nterms = int(input("შეიყვანოთ დადებითი მთელი რიცხვი "))

n1, n2 = 0, 1
count = 0

if nterms <= 0:
    print("გთხოვთ შეიყვანოთ დადებითი მთელი რიცხვი")
else:
    print("ფიბონაჩის მიმდევრობა:")
    while count < nterms:
        print(n1)
        nth = n1 + n2
        n1 = n2
        n2 = nth
        count += 1
```
